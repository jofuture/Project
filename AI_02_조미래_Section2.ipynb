```python
# elimbasu5 = 강가에 오물이나 쓰레기를 버리는지 여부 - 모든가구에서 No라고 응답했으므로 삭제
data.drop(['elimbasu5'],axis=1,inplace=True)

# 월세가 0인경우 집이 자가인지 확인하고 0으로 채워준다
data.loc[(data['tipovivi1'] == 1), 'v2a1'] = 0
# 그래도 949개의 결측값이 있다. -> 컬럼추가
data['v2a1-missing'] = data['v2a1'].isnull()
```

```python
#  개인이 19세 이상이고 7세 미만이고 누락된 연대는 0으로 설정
data.loc[((data['age'] > 19) | (data['age'] < 7)) & (data['rez_esc'].isnull()), 'rez_esc'] = 0

# 결측값 7~19세 사이 플래그 추가
data['rez_esc-missing'] = data['rez_esc'].isnull()

# 이상치는 변수의 최대값인 5로 설정
data[data['rez_esc'].isnull()]
data['rez_esc-missing'] = data['rez_esc'].isnull()
data.loc[data['rez_esc'] > 5, 'rez_esc'] = 5
```

```python
heads = data.loc[data['parentesco1'] == 1].copy()
heads.groupby('v18q')['v18q1'].apply(lambda x: x.isnull().sum())
#가구당 태블릿수를 나타내는 v18q1특성에서 Nan은 태블릿을 소유 하지않았음을 의미, 0으로 채워준다.
data['v18q1'] = data['v18q1'].fillna(0)
```

```python
# 데이터 불일치 수정 -

#v14a는 화장실 없음이 1 sanitario1은 화장실 있음이 1 abastaguano은 물 공급이 없는경우가 1 ->  0으로 바꿔주자
data.loc[(data.v14a ==  1) & (data.sanitario1 ==  1) & (data.abastaguano == 0), "v14a"] = 0
data.loc[(data.v14a ==  1) & (data.sanitario1 ==  1) & (data.abastaguano == 0), "sanitario1"] = 0

# 가장이 교육을 못 받았다고 응답하면 0으로 바꿔준다
data.loc[data['edjefa'] == "no", "edjefa"] = 0
data.loc[data['edjefe'] == "no", "edjefe"] = 0

# 가장이 교육을 받았다면 총 학교다닌년도인 escolari를 채워준다
data.loc[(data['edjefa'] == "yes") & (data['parentesco1'] == 1), "edjefa"] = data.loc[(data['edjefa'] == "yes") & (data['parentesco1'] == 1), "escolari"]
data.loc[(data['edjefe'] == "yes") & (data['parentesco1'] == 1), "edjefe"] = data.loc[(data['edjefe'] == "yes") & (data['parentesco1'] == 1), "escolari"]

data.loc[data['edjefa'] == "yes", "edjefa"] = 4
data.loc[data['edjefe'] == "yes", "edjefe"] = 4

# # 최대치로 교육받은 가장들에 대한 컬럼 추가
# data['edjef'] = np.max(data[['edjefa','edjefe']], axis=1)
```

```python
# 데이터타입 변환 및 값 치환
mapping = {"yes": 1, "no": 0}

# Apply same operation to both train and test
for df in [data]:
    # Fill in the values with the correct mapping
    data['dependency'] = data['dependency'].replace(mapping).astype(np.float64)
    data['edjefa'] = data['edjefa'].replace(mapping).astype(np.float64)
    data['edjefe'] = data['edjefe'].replace(mapping).astype(np.float64)

data[['dependency', 'edjefa', 'edjefe']].describe()
```

```python
# 새로운 전력 컬럼
data['hhsize-diff'] = data['tamviv'] - data['hhsize']
elec = []

for i, row in data.iterrows():
    if row['noelec'] == 1:#집에 전기가 안들어오는 가구
        elec.append(0)
    elif row['coopele'] == 1:
        elec.append(1)
    elif row['public'] == 1:
        elec.append(2)
    elif row['planpri'] == 1:
        elec.append(3)
    else:
        elec.append(np.nan)
        
data['elec'] = elec
data['elec-missing'] = data['elec'].isnull()

# 기존 전력 컬럼 제거
data = data.drop(columns = ['noelec', 'coopele', 'public', 'planpri'])
```

```python
# 가장이 표시된 가구
households_leader = data.groupby('idhogar')['parentesco1'].sum()
# 가장이 없는 가구
households_no_head = data.loc[data['idhogar'].isin(households_leader[households_leader == 0].index), :]
print('{}가구는 가장이없다..'.format(households_no_head['idhogar'].nunique()))

households_no_head_equal = households_no_head.groupby('idhogar')['Target'].apply(lambda x: x.nunique() == 1)
print('{} Households with no head have different labels.'.format(sum(households_no_head_equal == False)))
```

```python
!pip install --upgrade category_encoders
from sklearn.metrics import plot_confusion_matrix
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report
from sklearn.linear_model import LogisticRegression
from scipy.stats import randint, uniform
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler
from xgboost import XGBClassifier
from sklearn.metrics import f1_score
from category_encoders import OrdinalEncoder
from sklearn.pipeline import make_pipeline

from sklearn.impute import SimpleImputer 

pipe = make_pipeline(
    OrdinalEncoder(),
    SimpleImputer(),
    StandardScaler(),
    RandomForestClassifier(random_state=10,
                           n_estimators=100,
                           n_jobs=-1,
                           class_weight='balanced')#oob_score=True
)
pipe.fit(X_train, y_train)
print('검증 정확도: ', pipe.score(X_val, y_val))
```
